```python
#importing pandas the package
import pandas as pd
```

```python
#reading the fifa data into the dataframe df
df = pd.read_csv("/content/fifa.csv")
df.head()
```

```python
#setting the Name column as index for the dataframe
df.set_index("Name", inplace = True)
```

```python
#checking the dataset
df.head()
```

```python
#sorting the dataframe by age whuch is by default in ascending order
df.sort_values("Age")
```

```python
#checking for null values
df.isnull().sum()
```

```python
#statistical info on df
df.describe()
```

```python
df.info()
```

```python
#sorting age in descending order to see the oldest player
df.sort_values("Age", ascending = False).head()
```

```python
#players that are older than 30
older= df[df["Age"] > 30]
older.head()
```

```python
#players that are older than 30 and are high performing
older_better = df[(df["Age"] > 30) & (df["Overall"] > 90)]
older_better.head()
```

```python
#creating a new column
df["Rating per age"] = df["Overall"]/ df["Age"]
df["Rating per age"].head()
```

```python
#confirming that the column was added
df.columns
```

```python
#dropping the new column
df = df.drop("Rating per age", axis = 1)
```

```python
df.columns
```

```python
#mean performance based on age
df.groupby("Age").mean().head()
```

```python
#average age by nationality
avg_age = df.groupby("Nationality")["Age"].mean()
avg_age.head()
```
